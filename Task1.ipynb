```python
#importing pandas for data cleaning and preprocessing
import pandas as pd
#reading the csv file
df=pd.read_csv("netflix_titles.csv")
print(df)
```

```python
#Removing null values present in CSV file 
#inplace=True means null values currently present in CSV file is updated 
df.dropna(inplace=True)
```

```python
#displaying first 100 rows
df.head(100)
print(df.to_string())
```

```python
#Checking whether there are duplicates .If 'True' duplicates are present if 'False' none are there 
df.duplicated()
```

```python
#dropping duplicates
df.drop_duplicates()
```

```python
#converting column to lowercase and removing spaces if present to '_' to underscore
df.columns=[col.lower().replace(" ","_")for col in df.columns]
print(df)
```

```python
#Checking datatypes
df.dtypes
```

```python
#Converting each data to its own datatype
df['show_id'] = df['show_id'].astype(str)
df['type'] = df['type'].astype(str)
df['title'] = df['title'].astype(str)
df['director'] = df['director'].astype(str)
df['cast'] = df['cast'].astype(str)
df['country'] = df['country'].astype(str)
df['date_added'] = pd.to_datetime(df['date_added'])  # datetime
df['release_year'] = df['release_year'].astype(int)   # integer
df['rating'] = df['rating'].astype(str)
df['duration'] = df['duration'].astype(str)
df['listed_in'] = df['listed_in'].astype(str)
df['description'] = df['description'].astype(str)
```

```python
#Displaying its datatype after conversion
print(df.dtypes)
```
